#### **What is data?**
Data is a collection of facts, numbers, words, observations or other useful information. 
Through data processing and data analysis, organizations transform raw data points into valuable insights that improve decision-making and drive better business outcomes. For examples, 
- spreadsheet with sales information
- data generated on social media
- data generated by mobile applications
- website traffic data
- videos
- texts
- images
- geolocation data
- data generated by satellite sensors

#### **Types of Data**
- **Quantitative (Numeric):** Data that can be measured in numerical form. The values are expressed on a numeric scale. It is used to describe numeric variables.
    - **Continuous:** Data that can take on any value in an interval. It involves real numbers. For examples, \[-1, 1], \[0,1], etc.
    - **Discrete:** Data that can take on only interger values, such as counts. It involves only positive or negative numbers. For examples 1, 2, 5, 10, -4, etc..

- **Qualitative (Categorical):** Data that is collected in a non-numerical form. It is descriptive and involves text or categories. It can take on only a specific set of values representing a set of possible categories.
    - **Nominal:** Data that do not have a natural order or ranking. They cannot be measured. Calculations with these data are meaningless. For examples, gender, marital status, color, etc.
    - **Ordinal:** Categories that have a meaningful order or ranking but no consistent difference between levels. Order matters but not the difference between values. For examples, ratings of a products, socio economic class, education levels, etc.
    - **Binary:** A special type of categorical data which has just binary values. It can take only two values. It is also known as Dichotomous data. For examples, yes or no, True or False, 0 or 1, etc..

#### **Forms of Data**
- **Structured Data:** It is organized in a clear, defined tabular format, where each row is record and each column is variable. It can consist of both quantitative and qualitative data. The highly organized nature of structured data allows for quick querying and data analysis, making it useful for business intelligence systems and reporting processes.
  For examples, CSV, Spreadsheets, Relational Databases, etc

- **Unstructured Data:** It lacks a strictly defined format. It often comes in complex forms such as text documents, images and videos. Unstructured data can include both qualitative information (such as customer comments) and quantitative elements (such as numerical values embedded in text). For examples, emails, social media content and multimedia files.

- **Semi-structured Data:** Semi-structured data blends elements of structured and unstructured data. It doesn't follow a rigid format but can include tags or markers that make it easier to organize and analyze. It is widely used in scenarios such as web scraping and data integration projects because it offers flexibility while retaining some structure for search and analysis. For examples, XML files, JSON objects, etc.

- **Metadata:** Metadata is data about data. In other words, it is information about the attributes of a data point or data set, such as file names, authors, creation dates or data types. Metadata enhances data organization, searchability and management. It is critical to systems such as databases, digital libraries and content management platforms because it helps users more easily sort and find the data they need.

- **Big Data:** Big data refers to extremely large and complex data sets that cannot be easily managed or analyzed with traditional data processing tools, particularly spreadsheets. Big data includes structured data, unstructured data, mixed data sets, etc. It is recognized by three characteristics: **variety**, **volume**, and **velocity**, also known as the “three Vs.” In the past recent years, **veracity** and **value** is also added to the characteristics.
